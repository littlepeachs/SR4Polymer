```python
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
import xgboost as xgb
aging_data = pd.read_csv('./data/aging_mirco_macro.csv')
aging_data = aging_data.iloc[:, 1:]
aging_data = pd.DataFrame(scaler.fit_transform(aging_data), columns=aging_data.columns)
y = aging_data.iloc[:, -1]  # 
X = aging_data.iloc[:, :11]  # 

# # 
# train, test, labels_train, labels_test = train_test_split(x, y, train_size=0.8, random_state=2579)
# xgb_model = xgb.XGBRegressor(learning_rate=0.1, n_estimators=100)
# xgb_model.fit(train, labels_train)
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# 
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

#  PCA
pca = PCA(n_components=5)  # 95%
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# 
model = LinearRegression()
model.fit(X_train_pca, y_train)

# 
y_pred = model.predict(X_test_pca)

# 
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')

# 
print(f'Number of components: {pca.n_components_}')
print(":\n", pca.components_)
for i, component in enumerate(pca.components_):
    feature_contributions = np.abs(component)
    max_index = np.argmax(feature_contributions)
    print(f" {i+1}  {max_index} {feature_contributions[max_index]}")
```

```python
import xgboost
import shap
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
# train XGBoost model
random_seed = 42
aging_data = pd.read_csv('./data/aging_mirco_macro.csv')
aging_data = aging_data.iloc[:, 1:]
aging_data = pd.DataFrame(scaler.fit_transform(aging_data), columns=aging_data.columns)
y = aging_data.iloc[:, -1]  # 
X = aging_data.iloc[:, :11]  # 
model = xgboost.XGBRegressor(n_estimators=100, max_depth=2, random_state=random_seed).fit(X, y)
plt.figure(figsize=(12, 8))
# compute SHAP values
explainer = shap.Explainer(model, X)
shap_values = explainer(X[:1000])
plt.gca().set_facecolor('#FFFFFF')
shap.plots.heatmap(shap_values, show = False)

plt.savefig('./shap.png',dpi=600)
```

```python

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np

# 
X, y = load_breast_cancer(return_X_y=True)

# 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# 
rf = RandomForestClassifier(n_estimators=100, random_state=1)
rf.fit(X_train, y_train)

# 
base_acc = accuracy_score(y_test, rf.predict(X_test))

# 
importances = []

# 
for i in range(X_train.shape[1]):
    X_temp = np.delete(X_train, i, axis=1)  # 
    rf.fit(X_temp, y_train)  # 
    acc = accuracy_score(y_test, rf.predict(np.delete(X_test, i, axis=1)))  # 
    importances.append(base_acc - acc)  # 

# 
plt.figure(figsize=(10, 6))
plt.bar(range(len(importances)), importances, color='skyblue')
plt.xlabel('')
plt.ylabel('')
plt.title('Leave-one-out ')
plt.xticks(range(len(importances)), range(1, len(importances) + 1))
plt.show()
```
